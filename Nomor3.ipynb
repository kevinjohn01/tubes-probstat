```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

anggur = pd.read_csv("anggur.csv")
```

```python
#Normality Test dapat dilakukan dengan fungsi normaltest yang disediakan library scipy untuk memperoleh nilai statistic dan P-value
def NormalityTest(column):
    statistic,pValue = stats.normaltest(anggur[column])
    alpha = 0.05
    print("Nilai statistic:", statistic)
    print("p-Value:", pValue)

    #Jika p-value lebih besar daripada alpha, maka data dapat dikatakan berdistribusi normal
    if (pValue < alpha):
        print(column.capitalize(), "tidak berdistribusi normal.") #H0 ditolak
    else:
        print(column.capitalize(), "berdistribusi normal.")
    return
```

```python
#Tes Distribusi Normal: Fixed Acidity
NormalityTest("fixed acidity")
```

```python
#Histogram Fixed Acidity
sns.histplot(anggur["fixed acidity"], kde=True)
#Keterangan: kde=True menampilkan garis hubung
plt.show()
```
