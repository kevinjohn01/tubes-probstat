```python
import matplotlib.pyplot as plt
import pandas as pd

anggur = pd.read_csv("anggur.csv")
```

```python
# Fixed Acidity
anggur["fixed acidity"].plot(kind = "hist",rwidth=0.8)
plt.show()
```

```python
plt.boxplot(anggur["fixed acidity"])
plt.show()
```

```python
%matplotlib inline
df = pd.DataFrame(data=anggur, columns=['fixed acidity'])
boxplot = df.boxplot(column=['fixed acidity'])
plt.show()
```
